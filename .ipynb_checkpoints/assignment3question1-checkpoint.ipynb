```python
# The following line will import LogisticRegression and DecisionTreeClassifier Classes
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

# Importing the required packages and libraries
# we will need numpy and pandas later
import numpy as np
import pandas as pd

# A. Read the data file “Hearts_s.csv” (from github using the following command), and assign it to a Pandas DataFrame:   
heart_disease_df = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Heart_s.csv")

```

```python
# B.Check out the dataset. As you see, the dataset contains a number of features including both contextual and biological factors (e.g. age, gender, vital signs, …). The last column “AHD” is the label with “Yes” meaning that a human subject has Heart Disease, and “No” meaning that the subject does not have Heart Disease.
#print the first 10
heart_disease_df
```

```python
# C. As you see, there are at least 3 categorical features in the dataset (Gender, ChestPain, Thal). Let’s ignore these categorical features for now, only keep the numerical features and build your feature matrix and label vector.
features_cols = ['Age', 'RestBP', 'Chol', 'RestECG', 'MaxHR', 'Oldpeak']


# this is our independent value our features
X = heart_disease_df[features_cols]
# this is our dependent value our label
y= heart_disease_df['AHD']


```

```python
# D. Split the dataset into testing and training sets with the following parameters: test_size=0.25, random_state=4.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 4)

#print the size of the training data should be .75 of 301
print(X_train.shape)
print(y_train.shape)
```

```python
# E. Use KNN (with k=3), Decision Tree (with random_state=5 (this random state is for decision tree and you put it when you define the decision tree classifier. It is different from the random state that you used to split the data in part D)), and Logistic Regression Classifiers to predict Heart Disease based on the training/testing datasets that you built in part (d). Then check, compare, and report the accuracy of these 3 classifiers. Which one is the best? Which one is the worst?
 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

k=3
heart_knn = KNeighborsClassifier(n_neighbors = k)
heart_knn.fit(X_train,y_train)
y_predict_knn = heart_knn.predict(X_test)

heart_dt = DecisionTreeClassifier(random_state = 5)
heart_dt.fit(X_train, y_train)
y_predict_dt = heart_dt.predict(X_test)

heart_log = LogisticRegression()
heart_log.fit(X_train, y_train)
y_predict_log = heart_log.predict(X_test)

score_knn = accuracy_score(y_test, y_predict_knn)
score_dt = accuracy_score(y_test, y_predict_dt)
score_log = accuracy_score(y_test, y_predict_log)

print(score_knn)
print(score_dt)
print(score_log)

```

```python
#F. Now, we want to use the categorical features as well! To this end, we have to perform a feature engineering process called OneHotEncoding for the categorical features. To do this, each categorical feature should be replaced with dummy columns in the feature table (one column for each possible value of a categorical feature), and then encode it in a binary manner such that only one of the dummy columns can take “1” at a time (and zero for the rest). For example, “Gender” can take two values “m” and “f”. Thus, we need to replace this feature (in the feature table) by 2 columns titled “m” and “f”.  Wherever we have a male subject, we can put “1” and ”0” in the columns “m” and “f”.  Wherever we have a female subject, we can put “0” and ”1” in the columns “m” and “f”. (Hint: you will need 4 columns to encode “ChestPain” and 3 columns to encode “Thal”).


heart_disease_df_modified = heart_disease_df
heart_disease_df_modified

#the plan here is to copy the column and its data and then duplicate it to the categories we need after that we will remove the old column 

#gender_column will now contain the date from 'Gender'
gender_column = heart_disease_df[["Gender"]]
#adding 'Male' feature and 'Female feature into the datafram'
heart_disease_df_modified['Male'] = gender_column
heart_disease_df_modified['Female'] = gender_column
#this removes the 'Gender' feature
heart_disease_df_modified = heart_disease_df_modified.drop('Gender', 1)

#repeat the same thing as we did above but now for the ChestPain feature
chest_pain_column = heart_disease_df[["ChestPain"]]
heart_disease_df_modified['Typical'] = chest_pain_column
heart_disease_df_modified['NonTypical'] = chest_pain_column
heart_disease_df_modified['Asymptomatic'] = chest_pain_column
heart_disease_df_modified['Nonanginal'] = chest_pain_column
heart_disease_df_modified = heart_disease_df_modified.drop('ChestPain', 1)

#repeat the same thing as we did above but now for the Thal feature
thal_column = heart_disease_df[["Thal"]]
heart_disease_df_modified['Fixed'] = thal_column
heart_disease_df_modified['Normal'] = thal_column
heart_disease_df_modified['Reversable'] = thal_column
heart_disease_df_modified = heart_disease_df_modified.drop('Thal', 1)

#for the Gender features we broken into two columns
heart_disease_df_modified['Male'] = heart_disease_df_modified['Male'].replace(['m'], '1')
heart_disease_df_modified['Male'] = heart_disease_df_modified['Male'].replace(['f'], '0')

heart_disease_df_modified['Female'] = heart_disease_df_modified['Female'].replace(['m'], '0')
heart_disease_df_modified['Female'] = heart_disease_df_modified['Female'].replace(['f'], '1')

#for the ChestPain we broken into four columns
heart_disease_df_modified['Typical'] = heart_disease_df_modified['Typical'].replace(['typical'], '1')
heart_disease_df_modified['Typical'] = heart_disease_df_modified['Typical'].replace(['nontypical'], '0')
heart_disease_df_modified['Typical'] = heart_disease_df_modified['Typical'].replace(['nonanginal'], '0')
heart_disease_df_modified['Typical'] = heart_disease_df_modified['Typical'].replace(['asymptomatic'], '0')

heart_disease_df_modified['NonTypical'] = heart_disease_df_modified['NonTypical'].replace(['typical'], '0')
heart_disease_df_modified['NonTypical'] = heart_disease_df_modified['NonTypical'].replace(['nontypical'], '1')
heart_disease_df_modified['NonTypical'] = heart_disease_df_modified['NonTypical'].replace(['nonanginal'], '0')
heart_disease_df_modified['NonTypical'] = heart_disease_df_modified['NonTypical'].replace(['asymptomatic'], '0')

heart_disease_df_modified['Asymptomatic'] = heart_disease_df_modified['Asymptomatic'].replace(['typical'], '0')
heart_disease_df_modified['Asymptomatic'] = heart_disease_df_modified['Asymptomatic'].replace(['nontypical'], '0')
heart_disease_df_modified['Asymptomatic'] = heart_disease_df_modified['Asymptomatic'].replace(['nonanginal'], '0')
heart_disease_df_modified['Asymptomatic'] = heart_disease_df_modified['Asymptomatic'].replace(['asymptomatic'], '1')

heart_disease_df_modified['Nonanginal'] = heart_disease_df_modified['Nonanginal'].replace(['typical'], '0')
heart_disease_df_modified['Nonanginal'] = heart_disease_df_modified['Nonanginal'].replace(['nontypical'], '0')
heart_disease_df_modified['Nonanginal'] = heart_disease_df_modified['Nonanginal'].replace(['nonanginal'], '1')
heart_disease_df_modified['Nonanginal'] = heart_disease_df_modified['Nonanginal'].replace(['asymptomatic'], '0')

#for the thal column we broken into three columns
heart_disease_df_modified['Fixed'] = heart_disease_df_modified['Fixed'].replace(['fixed'], '1')
heart_disease_df_modified['Fixed'] = heart_disease_df_modified['Fixed'].replace(['normal'], '0')
heart_disease_df_modified['Fixed'] = heart_disease_df_modified['Fixed'].replace(['reversable'], '0')

heart_disease_df_modified['Normal'] = heart_disease_df_modified['Normal'].replace(['fixed'], '0')
heart_disease_df_modified['Normal'] = heart_disease_df_modified['Normal'].replace(['normal'], '1')
heart_disease_df_modified['Normal'] = heart_disease_df_modified['Normal'].replace(['reversable'], '0')

heart_disease_df_modified['Reversable'] = heart_disease_df_modified['Reversable'].replace(['fixed'], '0')
heart_disease_df_modified['Reversable'] = heart_disease_df_modified['Reversable'].replace(['normal'], '0')
heart_disease_df_modified['Reversable'] = heart_disease_df_modified['Reversable'].replace(['reversable'], '1')

#we are going to move AHD feature to last column because its also going to be our dependent value
ahd_column = heart_disease_df_modified.pop('AHD')
heart_disease_df_modified.insert(15, 'AHD', ahd_column);

heart_disease_df_modified 
```

```python
#G. Repeat parts (d) and (e) with the new dataset that you built in part (f). How does the prediction accuracy change for each method?

new_features_cols = ['Age', 'RestBP', 'Chol', 'RestECG', 'MaxHR', 'Oldpeak', 'Male', 'Female', 'Typical', 'NonTypical', 'Asymptomatic', 'Nonanginal', 'Fixed', 'Normal', 'Reversable']

# this is our independent value our features
X_modified = heart_disease_df_modified[new_features_cols]
# this is our dependent value our label
y_modified = heart_disease_df_modified['AHD']


X_train_modified, X_test_modified, y_train_modified, y_test_modified = train_test_split(X_modified, y_modified, test_size = 0.25, random_state = 4)

k=3
heart_knn_modified = KNeighborsClassifier(n_neighbors = k)
heart_knn_modified.fit(X_train_modified,y_train_modified)
y_predict_knn_modified = heart_knn_modified.predict(X_test_modified)

heart_dt_modified = DecisionTreeClassifier(random_state = 5)
heart_dt_modified.fit(X_train_modified, y_train_modified)
y_predict_dt_modified = heart_dt_modified.predict(X_test_modified)

heart_log_modified = LogisticRegression()
heart_log_modified.fit(X_train_modified, y_train_modified)
y_predict_log_modified = heart_log_modified.predict(X_test_modified)

score_knn_modified = accuracy_score(y_test_modified, y_predict_knn_modified)
score_dt_modified = accuracy_score(y_test_modified, y_predict_dt_modified)
score_log_modified = accuracy_score(y_test_modified, y_predict_log_modified)

print(score_knn_modified)
print(score_dt_modified)
print(score_log_modified)

#For KNN with the extra features remained the same and for the Decision Tree and Logistic Regression the accuray was higher.


```

```python
# H. Now, repeat part (e) with the new dataset that you built in part (f), but this time using Cross-Validation. Thus, rather than splitting the dataset into testing and training, use 10-fold Cross-Validation (as we learned in Lab4) to evaluate the classification methods and report the final prediction accuracy. 

from sklearn.model_selection import cross_val_score

#applying 10 fold cross validation for "logistic regression"
#225 of rows is the number of our training data
heart_disease_logreg_modified = LogisticRegression(max_iter = 301)

accuracy_list_logreg = cross_val_score(heart_disease_logreg_modified, X_modified, y_modified, cv=10, scoring ='accuracy')

print(accuracy_list_logreg)

accuracy_cv_logreg = accuracy_list_logreg.mean()
print(accuracy_cv_logreg)
```
